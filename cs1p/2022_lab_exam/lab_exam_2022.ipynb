### The similarity algorithm

If `user_a` enjoyed reading a number of books that `user_b` also enjoyed reading, then when `user_a` recommends another book that `user_b` has not read, chances are that `user_b` would like to read it too. On the other hand, if the two users always tend to disagree about books, then `user_b` is unlikely to read a book that `user_a` recommends. 

You can calculate how similar two users are by treating each of their ratings as a vector and calculating the dot product of these two vectors. The dot product is simply the sum of the products of each of the corresponding elements.

For example, suppose we have 3 books in our database and

* `user_a` rated them `[5, 3, -5]` (here 5, 3, and -5 is `user_a`'s rating for books 1, 2 and 3 respectively),
* `user_b` rated them `[1, 5, -3]`, 
* `user_c` rated them `[5, -3, 5]`, and 
* `user_d` rated them `[1, 3, 0]`. 

The similarity between `user_a` and `user_b` is calculated as: 

$$(5 \times 1) + (3 \times 5) + (-5 \times -3) = 5 + 15 + 15 = 35.$$

The similarity between `user_a` and `user_c` is: 

$$(5 \times 5) + (3 \times -3) + (-5 \times 5) = 25 - 9 - 25 = -9.$$

The similarity between `user_a` and `user_d` is 

$$(5 \times 1) + (3 \times 3) + (-5 \times 0) = 5 + 9 + 0 = 14.$$

We see that if two people like a book (rating it with a positive number), it increases their similarity; and if two people dislike a book (both giving it a negative number), it also increases their similarity.

Once you have calculated the pair-wise similarity between `user_a` and every other user, you can then identify whose ratings are most similar to `user_a`'s. In the above case, `user_a` is most similar to `user_b`, so we would recommend to `user_a` the top books from `user_b`'s list that `user_a` hasn't already read.

```python
################################ FUNCTIONS #################

# Returns dot product of two lists.
def dot_product(lst1, lst2):
    zipped = list(zip(lst1, lst2))
    dots = [i[0] * i[1] for i in zipped]
    return sum(dots)

# Reads in rating and book filenames, returns ratings and books as a dictionary of lists and list of tuples.
def read_in(ratings_filename, books_filename):
    name = True
    all_names = []
    all_ratings = []
    books = []
    
    # produces two seperate lists, they puts them together in dictionary
    with open(ratings_filename) as f:
        for line in f:
            if name == True:
                all_names.append((line.strip()).title())
            else:
                all_ratings.append(list(map(lambda a: int(a), (line.strip()).split(" "))))
            name = not name
        ratings = {name:all_ratings[index] for index, name in enumerate(all_names)}
        
    with open(books_filename) as f:
        for line in f:
            book = tuple((line.strip()).split(","))
            books.append(book)
    
    return ratings, books

# sorts in reverse names in tie, returns appropriate name
def resolve_tie(score_lst, main_score):
    names_found = []
    for pair in score_lst:
        if pair[1] == main_score:
            names_found.append(pair[0])
    sorted_names = sorted(names_found, reverse = True)
    return sorted_names[0]
                          
# Generates the message to be displayed.
def generate():
    
    try:
        # validate variables
        ratings, books = read_in("ratings.txt", "books.txt")
        name = (name_field.get()).title()
        user_ratings = ratings[name]
        recommendation_number = recommend_field.get()
        
        # validate the input
        recommendation_number = int(recommendation_number)
        if recommendation_number <= 0:
            raise IndexError("Please enter a number greater than 0.")
         
    # tell user what the error is
    except ValueError:
        message["text"] = f"'{recommendation_number}' isn't a valid number."
        return
    except KeyError:
        message["text"] = f"There is no user called '{name}' in our database."
        return
    except FileNotFoundError:
        message["text"] = "Unable to import data, required files are missing."
        return
    except IndexError as error:
        message["text"] = error
        return
    
    # the text that'll be displayed, the program will add to this later
    seperator = "+++++++++++++++++++++++++++++++++++++++++++++++++++\n"
    recommendations_msg = f"\nYour recommendation based on the simiarity algorithm:\n{seperator}"
    
    # calculate scores for each person, find largest scores
    scores = []
    sorted_scores = []
    for other_person in ratings.keys():
        if name != other_person:
            scores.append((other_person, dot_product(user_ratings, ratings[other_person])))
            sorted_scores.append(scores[-1][1])
    sorted_scores.sort()

    books_found = 0 # tracks number of books already recommended
    people_already_covered = set() # tracks what people have already been looked over
    books_already_recommended = set() # tracks what books have already been recommended
    main_start = True # tracks whether a book has been found
    
    # loops until the amount of books recommeded is enough or there is no more people to check
    while books_found <= recommendation_number and len(sorted_scores) > 0:
        if len(sorted_scores) >= 2 and sorted_scores[-1] == sorted_scores[-2]:
            tie = True # tracks ties
        else:
            tie = False
        score = sorted_scores.pop()
        
        # find the similar person based off similarity score, use function to resolve ties
        if tie == False:
            for person in scores:
                if person[1] == score and person[0] not in people_already_covered:
                    similar_person = person[0]
                    people_already_covered.add(similar_person)
                    start = True # tracks whether "Books recommended by..." should be printed
                    break
        elif tie == True:
            similar_person = resolve_tie(scores, score)
            people_already_covered.add(similar_person)
            start = True # tracks whether "Books recommended by..." should be printed
        
        # finds books to recommend, prints results
        for index, rating_given in enumerate(ratings[similar_person]):
            if books_found >= recommendation_number:
                break
            elif (rating_given == 3 or rating_given == 5) and user_ratings[index] == 0 and start == True and main_start == True and books[index][1] not in books_already_recommended:
                recommendations_msg = recommendations_msg + f"\nCongratulations, we are able to recommend INSERT_NUMBER_FOUND books for you!\n\n"
                recommendations_msg = recommendations_msg + f"Recommending based on the similarity algorithm.\n" + seperator
                recommendations_msg = recommendations_msg + f"\nBooks recommended by {similar_person} with similarity {score}:\n"
                recommendations_msg = recommendations_msg + f"{books[index][1]}, written by {books[index][0]}\n"
                books_found += 1
                books_already_recommended.add(books[index][1])
                start = False
                main_start = False
            elif (rating_given == 3 or rating_given == 5) and user_ratings[index] == 0 and start == True and books[index][1] not in books_already_recommended:
                recommendations_msg = recommendations_msg + f"\nBooks recommended by {similar_person} with similarity {score}:\n"
                recommendations_msg = recommendations_msg + f"{books[index][1]}, written by {books[index][0]}\n"
                books_found += 1
                books_already_recommended.add(books[index][1])
                start = False
            elif (rating_given == 3 or rating_given == 5) and user_ratings[index] == 0 and books[index][1] not in books_already_recommended:
                recommendations_msg = recommendations_msg + f"{books[index][1]}, written by {books[index][0]}\n"
                books_found += 1
                books_already_recommended.add(books[index][1])
    
    # finish message formatting, send results to user
    recommendations_msg = recommendations_msg + f"\n{seperator}"
    final_msg = recommendations_msg.replace("INSERT_NUMBER_FOUND", f"{books_found}")
    message["text"] = final_msg
    return

# resets input fields and message
def reset():
    name_field.delete(0,"end")
    recommend_field.delete(0,"end")
    message["text"] = "Recommendations will be displayed here."
    return

##################### G.U.I ####################################
        
root = tk.Tk()
root.title("Book recommendation system")

output_frame = tk.Frame(root, relief = tk.SUNKEN, borderwidth=2)
output_frame.grid(row=10,rowspan=5,columnspan=10,pady=10,padx=10,sticky="ew")

message = tk.Label(output_frame, text = "Recommendations will be displayed here.")
message.pack()

introduction = tk.Label(root, text = "Enter username and number of book recommendations below")
introduction.grid(row=0,column=1,columnspan=9,padx=5,pady=5)

# This is where the user inputs their name
name_label = tk.Label(root, text = "What is your name?")
name_label.grid(row=1,column=0,sticky="e")
name_field = tk.Entry(root)
name_field.grid(row=1,column=1,columnspan=9,sticky="ew")

# This is where the user inputs the number of recommendations
recommend_label = tk.Label(root, text = "How many recommendations?")
recommend_label.grid(row=2,column=0,sticky="e")
recommend_field = tk.Entry(root)
recommend_field.grid(row=2,column=1,columnspan=9,sticky="ew")

generate_btn = tk.Button(root, text = "Submit", bg = "green", fg = "white", command = generate)
generate_btn.grid(row=6,column=4,pady=5,padx=5)

reset_btn = tk.Button(root, text = "Reset", bg = "red3", fg = "white", command = reset)
reset_btn.grid(row=6,column=5,pady=5,padx=5)

quit_btn = tk.Button(root, text = "Quit", bg = "red", fg = "white", command = root.destroy)
quit_btn.grid(row=6,column=6,pady=5,padx=5)

root.mainloop()
```

The linearithmic time comes from the 'resolve_tie' function, which uses the sorted() function on a list of users with the same similarity scores. Although it is unlikely that everyone would share the same similarity score, the worst case is O(nlogn).

The worst case time complexity for the total number of books (assuming user ratings increase with the number of books) is O(m^2). This is due to a for loop being nested within a while loop. The amount of times the while loop loops is partially based off the variable 'books_found' which is incremented in a loop that is based off the number of book ratings, giving it a complexity of O(m).
